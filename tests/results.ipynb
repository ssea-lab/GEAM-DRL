```python
#对于每个data里的每一行，按\t分割，第一个值是模型名，后面的值是数据，存到numpy里，列名分别为reward, user, costs, capacity, experience, server
# 最终就变成10 * 5 * 6的三维表格
# 先创建好三维的numpy
import numpy as np
result = np.zeros((10, 6, 6))

for i, data in enumerate(datas):
    for j, line in enumerate(data):
        line = line.split("\t")
        result[i, j] = np.array(line[1:], dtype=float)
```
